### A Script to check that response removal has occurred correctly

###### Script only works if inputted waveform file is in the local directory.

###### Script should work for both recently downloaded, corrected waveform data, and older files pulled manually from eqserver.

Idea of this script is to attatch the metadata file UOM.xml, as pulled down from our GitHub page, remove the instrument response and plot the data, whilst also listing the maximum values. If everything as working as it should, these values will be quite comparible to the max values given by eqserver. 

A good example event to download and use is the ML 2.2 on 02/12/2014 at UTC 1634. This is the file used in testing this code and finalising the metadata corrections. All stations give max values close to those provided by eq server, apart from S88U, which is at a significant distance from the event compared to other stations.

```python
from obspy.station import stationxml
from obspy import read, read_inventory
import os 
import numpy
import matplotlib
%matplotlib inline


#import and read metadata file from GitHub
os.system('wget https://raw.githubusercontent.com/unimelb-geophys/metadata/master/UOM.xml')



waveform = raw_input("enter waveform file name, e.g. 2014-12-2UTC16_33.mseed:")

inv = stationxml.read_StationXML('UOM.xml')
st = read(waveform)



##COrrections copied from other code
x = len(st)

network_change1 = 'UM'
network_new_name1 = 'BM'
network_change2 = '01'
network_new_name2 = 'UM'
network_change3 = ''
network_new_name3 = 'UM'
location_blank = ''


for i in range(0, x):
    tr = st[i]
    
    
# removes LOCATION so it is blank, as listed in the metadata files (regardless of what it was previously)
   
    tr.stats["location"] = location_blank
    
# Changes BOREHOLE network codes from UM to BM and SURFACE network codes from 01 to UM 
    net = tr.stats["network"]
    if network_change1 in net:
        tr.stats["network"] = network_new_name1
    elif network_change2 in net:
        tr.stats["network"] = network_new_name2
    elif network_change3 in net:
        tr.stats["network"] = network_new_name3
    else:
        continue

#CORRECTING BOREHOLE STATION NAMES

serial_no_1 = 'A346'
site_name_1 = 'LOYU'
serial_no_2 = 'BD5E'
site_name_2 = 'MOSU'
serial_no_3 = 'BD70'
site_name_3 = 'SGWU'
serial_no_4 = 'BD91'
site_name_4 = 'WILU'

#Changes station name from serial number to station code

for i in range(0, x):
    tr = st[i]
    stat = tr.stats["station"] 
    if serial_no_1 in stat:
        tr.stats["station"] = site_name_1
    elif serial_no_2 in stat:
        tr.stats["station"] = site_name_2
    elif serial_no_3 in stat:
        tr.stats["station"] = site_name_3
    elif serial_no_4 in stat:
        tr.stats["station"] = site_name_4
    else:
        continue
        
# CHANGES TO CHANNEL CODE

# (this is a bit messy at the moment since the wildcard feature seemed to be failing)
        
channel_new_name_E = 'EHE'
channel_new_name_N = 'EHN'
channel_new_name_Z = 'EHZ'

for i in range(0, x):
    tr = st[i]
    chan = tr.stats["channel"] 

# Changes CHANNEL names from '**E', '**N', '**Z', (e.g. BHE, DHZ) to a consitant format of EHE, EHN, EHZ
# EXCEPT FOR BOREHOLE STATIONS, which will maintain channel codes BHE, BHN, BHZ

    if 'DHE' in chan:
        tr.stats["channel"] = channel_new_name_E
    elif 'DHN' in chan:
        tr.stats["channel"] = channel_new_name_N
    elif 'DHZ' in chan:
        tr.stats["channel"] = channel_new_name_Z    
    elif 'ENE' in chan:
        tr.stats["channel"] = channel_new_name_E
    elif 'ENN' in chan:
        tr.stats["channel"] = channel_new_name_N
    elif 'ENZ' in chan:
        tr.stats["channel"] = channel_new_name_Z         
    else:
        continue        
 

# saves stream as a combination of edited traces       
    st[i] = tr
    
#attach metadata, remove response and plt waveform, listing max values
    
st.attach_response(inv)
st.remove_response()
print st.plot()
print 'Max values for each trace are:'
st.max()
```

###### If everything has gone to plan, these values should be similar (or at leat within an order of magnitude) of those listed on eqserver.
